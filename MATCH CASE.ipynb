# Match Case


```python
# OPÇÃO ÚNICA

operacao = input("""
ESCOLHA UMA OPÇÃO:


[1] Opção 1
[2] Opção 2
[3] Opção 3


""")


match operacao: # match corresponde a operacao
    case '1':
        print("Opção 1 foi escolhida...")
    case '2':
        print("Opção 2 foi escolhida...")
    case '3':
        print("Opção 3 foi escolhida...")
    case _:
        print("Opção inválida...")
```

```python
# OPÇÃO MÚLTIPLA

operacao = input("""
ESCOLHA UMA OPÇÃO:


[1] Opção 1
[2] Opção 2
[3] Opção 3
""")


match operacao:
    case '1' | '2': # if operacao == '1' or operacao == '2':  O sinal | signfica "or"
        print("Opção 1 ou 2 escolhida...")
    case '3':
        print("Opção 3 foi escolhida...")
    case _:
        print("Opção inválida...")
```

```python
# OPÇÃO MÚLTIPLA COM CASE _ IF

numero = int(input("Digite um número: "))


match numero:
    case 1 | 3 | 5 | 7 | 9:
        print("É ímpar entre 0 e 10")
    case _ if numero < 0:
        print("É negativo")
    case _:
        print("Um número qualquer...")
```

```python
# DESAFIO 01
"""
Escreva um programa em Python que solicite ao usuário uma nota inteira entre 0 e 10.
Use o comando `match case` para classificar a nota de acordo com a tabela a seguir:


- 0-4: Nota Baixa
- 5-7: Nota Média
- 8-9: Nota Alta
- 10: Nota Excelente


Exiba a classificação correspondente.
"""

match input("\nDigite sua nota na última prova: \n\n"):
    case '10':
        print(f"\nTemos um prodígio, nota excelente! ")
    case '8' | '9':
        print(f"\nParabéns sua nota está muito boa! ")
    case '5' | '6'| '7':
        print(f"\nNão está ruim, mas da para melhorar ")
    case '0' | '1' | '2' | '3' | '4':
        print(f"\nPrecisa melhorar... ")
    case _:
        print("\nNota inválida")
```

```python
# DESAFIO 02.2
"""
Refaça o DESAFIO 01 considerando notas com casas decimais.
"""

match float(input("\nDigite sua nota na última prova: \n\n")):
    case 10:
        print(f"\nTemos um prodígio, nota excelente! ")
    case num if num >=8 and num <=9:
        print(f"\nParabéns sua nota está muito boa! ")
    case num if num >=5 and num <=7:
        print(f"\nNão está ruim, mas da para melhorar ")
    case num if num >=0 <=4:
        print(f"\nPrecisa melhorar... ")
    case _:
        print("\nNota inválida")
```

```python
# DESAFIO 04
"""
Implemente uma calculadora simples que realiza as operações de adição, subtração, multiplicação e divisão.
O usuário deve inserir dois números e, em seguida, selecionar a operação desejada (+, -, *, /).
Use o comando match case para realizar a operação correta e exibir o resultado.
"""

from google.colab import output

valor_1 = float(input("""Digite o número da calculadora

[1][2][3]
[4][5][5]
[7][8][9]
   [0]

"""))

output.clear()

operacao_aritimetica = input("""\nSelecione a opção:
[ + ]
[ - ]
[ * ]
[ / ]

""")

output.clear()

valor_2 = float(input("""Digite outro número da calculadora

[1][2][3]
[4][5][5]
[7][8][9]
   [0]

"""))

output.clear()

match operacao_aritimetica:
    case '+':
        print(f"\n{valor_1 + valor_2:.0f}")
    case '-':
        print(f"\n{valor_1 - valor_2:.0f}")
    case '*':
        print(f"\n{valor_1 * valor_2:.0f}")
    case '/':
        print(f"\n{valor_1 / valor_2:.0f}")
    case _:
        print("\nOperação inválida ❌")
```

```python
#DESAFIO 4.2

from google.colab import output

valor_1 = float(input("""Digite o número da calculadora

[1][2][3]
[4][5][5]
[7][8][9]
   [0]

"""))

output.clear()

operacao_aritimetica = input("""\nSelecione a opção:
[ + ]
[ - ]
[ * ]
[ / ]

""")

output.clear()

valor_2 = float(input("""Digite outro número da calculadora

[1][2][3]
[4][5][5]
[7][8][9]
   [0]

"""))

output.clear()

match operacao_aritimetica:
    case '+':
        print(f"\n{valor_1} + {valor_2} = {valor_1 + valor_2}")
    case '-':
        print(f"\n{valor_1} - {valor_2} = {valor_1 + valor_2}")
    case '*':
        print(f"\n{valor_1} * {valor_2} = {valor_1 + valor_2}")
    case '/':
        if valor_1 != '0' or valor_2 != '0':
            print(f"\n{valor_1} / {valor_2} = {valor_1 + valor_2}")
        else:
            print("Operação inválida, não existe divisão por zero ❌")
    case _:
        print("\nOperação inválida ❌")
```

```python
# DESAFIO 06
"""
Crie um programa que peça ao usuário para inserir uma letra. O programa deve utilizar match case
para verificar se a letra é uma vogal (a, e, i, o, u) ou uma consoante, e então exibir uma mensagem apropriada.
"""

caracter = input("Insira uma letra, para saber se é vogal, consoante, númerico ou não: ").lower()

match caracter:
    case 'a' | 'e' | 'i' | 'o' | 'u' :
        print("\nDefinitivamente é uma vogal")
    case _ if caracter.isalpha():
        print("\nÉ uma consoante")
    case _ if caracter.isnumeric():
        print("\nÉ caracter numérico")
    case _:
        print("\nÉ um caracter especial")
```

```python
# DESAFIO 06
"""
Crie um programa que peça ao usuário para inserir uma letra. O programa deve utilizar match case
para verificar se a letra é uma vogal (a, e, i, o, u) ou uma consoante, e então exibir uma mensagem apropriada.
"""

caracter = input("Insira uma letra, para saber se é vogal, consoante, númerico ou não: ").lower()

if len(caracter) == 1:
    match caracter:
        case 'a' | 'e' | 'i' | 'o' | 'u' :
            print("\nDefinitivamente é uma vogal")
        case _ if caracter.isalpha():
            print("\nÉ uma consoante")
        case _ if caracter.isnumeric():
            print("\nÉ caracter numérico")
        case _:
            print("\nÉ um caracter especial")
else :
    print("\nDigite apenas 1 letra")
```

```python
# DESAFIO 06 v3 (com função is float do lindo professor Leandro Toniati!)


def isfloat(numero):
    numero = str(numero)
    sep_numero = numero.split('.')
    if len(sep_numero) == 2 and sep_numero[0].isnumeric() and sep_numero[1].isnumeric():
        return True
    else:
        return False


letra = input("Digite uma letra: ").lower()
match letra:
    case 'a'|'e'|'i'|'o'|'u':
        print("\nÉ uma vogal!")
    case _ if letra.isalpha():
        print("\nÉ uma consoante!")
    case _ if letra.isnumeric():
        print("\nÉ um número inteiro!")
    case _ if isfloat(letra):
        print("\nÉ um número decimal!")
    case _:
        print("\nÉ um caractere especial!")
```
