```python
# DESAFIO 05
"""
Crie um programa que dados o valor, a taxa e o tempo, efetuar
o cálculo do valor de uma prestação em atraso.
FÓRMULA: valor da prestação + (valor da prestação * (taxa / 100) * tempo)
"""

# Considerei que o "tempo" seria no caso, mensalidades

taxa = 2 # Taxa fixa 2% ao mês

valor_prestacao = float(input("Qual o valor da prestação acordado anteriormente? \n"))

tempo_financiamento = float(input("\nQual o tempo, em meses, a parcela está atrasada? \n"))

prestacao_atrasada = valor_prestacao + (valor_prestacao * (taxa / 100) * tempo_financiamento)

val_anoatrasado = prestacao_atrasada * 12

print(f"\nConsiderando que possuimos uma taxa fixa de 2% ao mês para atrasos de prestação, por mês, sua prestação irá ficar em R${prestacao_atrasada:,.1f},"
      f" o valor total anual, ficou R${val_anoatrasado:,.1f}. \n")

print("Como o senhor(a) gostaria de pagar as prestações atrasadas?")
```
