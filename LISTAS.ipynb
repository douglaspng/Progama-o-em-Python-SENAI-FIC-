```python
# MANIPULAÇÃO DE LISTAS


bancos = ['Banco do Brasil', 'CEF', 'Santander']
print(bancos)
print(bancos[1])


# SUBSTITUIR ITEM DA LISTA
bancos[1] = 'Itaú'
print(bancos)


# SUBSTITUIR ÚLTIMO ITEM DA LISTA
bancos[-1] = 'C6'
print(bancos)


# AGREGAR ITENS A LISTA
bancos = bancos + ['Safra']
print(bancos)


# AGREGAR ITENS A LISTA (Operador de atribuição)
bancos += ['Bradesco', 'Nubank']
print(bancos)
```

```python
# MÉTODOS PARA LISTAS


lista = [4, 5, 3, 5]
print(lista)


# ADICIONAR ITENS A LISTA
lista.append(2)
print(lista)


# ADICIONAR VALOR NO ÍNDICE ESPECÍFICO
lista.insert(2, -3) # insert(índice, valor)
print(lista)


# CONTAR VALORES
print(lista.count(5))


# VERIFICAR TAMANHO DA LISTA
print(len(lista))


# OBTER ÍNDICE DO ITEM DA LISTA
print(lista.index(5))


# REVERTER ITENS DA LISTA
lista.reverse() # Não atribuir a uma variável
print(lista)


# ORDENAR A LISTA (A/Z)
lista.sort() # Não atribuir a uma variável
print(lista)


# ORDENAR A LISTA (Z/A)
lista.sort(reverse=True) # Não atribuir a uma variável
print(lista)


# REMOVER ITEM ATRAVÉS DO VALOR
lista.remove(5)
print(lista)


# REMOVER ÚLTIMO ITEM DA LISTA (recebe valor removido)
removeu = lista.pop() # remove pelo índice
print(removeu)
print(lista)


# REMOVER INTERVALO OU ÍNDICE
del lista[1:3]
print(lista)


# LIMPAR LISTA
lista.clear()
print(lista)


#del lista # Remove o objeto lista da memória
```

```python
# DESAFIO 01
"""
Faça um programa que leia 5 valores numéricos e
guarde-os em uma lista. No final, mostre qual foi
o maior e o menor valor digitado e as suas respectivas posições na lista.
"""

maior = ("-inf")
menor = ("inf")
valores = []

for i in range(1, 6):
    valor = input(f"Digite o {i}º valor ")
    valores.append(valor)

    maior = max(maior, valor)
    menor = min(menor, valor)

print(f"\nOs valores apresentados foram esses {valores}")
print(f"\nO maior valor apresentado foi {maior} e a posição dele foi a {valores.index(max(valores))+1}º")
print(f"\nO menor valor apresentado foi {menor} e a posição dele foi a {valores.index(min(valores))+1}º")



```

```python
# DESAFIO 02
"""
Crie um programa onde o usuário possa digitar vários valores
numéricos e cadastre-os em uma lista. Caso o número já exista
lá dentro, ele não será adicionado. No final serão exibidos
todos os valores únicos digitados, em ordem crescente.
"""

valores_lista = []

while True:
    numeros = int(input("\nDigite quantos valores você prefirir "))

    if numeros == 0:
        break

    if numeros in valores_lista:
        print("Número já cadastrado")
        continue

    valores_lista.append(numeros)

valores_lista.sort()
print(f"\nSua lista inteira de valores foi", *valores_lista)
```

```python
# DESAFIO 03
"""
Crie um programa que vai ler vários números e colocar em uma lista.
Depois disso, mostre:
A) Quantos números foram digitados.
B) A lista de valores, ordenada de forma decrescente.
C) Se o valor 5 foi digitado e está ou não na lista.
"""

lista = []

while True:
    numeros = int(input("Digite quantos números foram necessários, digite 0 para sair "))

    if numeros == 0:
        break

    lista.append(numeros)

if 5 in lista:
    print("\nO 5 está na sua lista")

lista.sort(reverse=True)

print(f"\nA Quantidade de números dígitados foram {len(lista)}")
print(f"\nSua lista inteira de valores foi,", *lista)
```

```python
# DESAFIO 04
"""
Crie um programa que vai ler vários números e colocar em uma lista.
Depois disso, crie duas listas extras que vão conter apenas os
valores pares e os valores ímpares digitados, respectivamente.
Ao final, mostre o conteúdo das três listas geradas.
"""


#Exercício finalizado, porém falta refinar para deixar mais "bonito"

todos_valores = []
valores_impares = []
valores_pares = []

while True:
    valores = int(input("Digite vários valores, 0 para sair "))

    todos_valores.append(valores)

    if valores%2 == 0:
        valores_pares.append(valores)

    if valores%2 == 1:
        valores_impares.append(valores)

    if valores == 0:
        break

todos_valores.sort()
valores_impares.sort()
valores_pares.sort()

print(f"Esses são todos os valores ímpares que você digitou {valores_impares}")
print(f"Agora esses aqui, são todos os valores pares {valores_pares}")
print(f"E por fim, aqui são todos os valores digitados {todos_valores}")
```

```python
# DESAFIO 05 BÔNUS
"""
Dada a lista com 32 Times, faça o sorteio para outras duas listas (A e B).
Ao final, cada nova lista deve conter 16 times.
random.choice(times)
"""
from random import choice

times=['Flamengo', 'Palmeiras', 'São Paulo', 'Athletico-PR', 'Atlético-MG', 'Corinthians', 'Fluminense', 'Grêmio',
       'Fortaleza', 'Internacional', 'Bahia', 'Botafogo', 'Red Bull Bragantino', 'Atlético-GO', 'Ceará', 'Cuiabá',
       'Goiás', 'Vasco', 'Juventude', 'Sport', 'CRB', 'Vitória', 'Criciúma', 'Sampaio Corrêa', 'Operário-PR',
       'Botafogo-SP', 'Brusque', 'Ypiranga-RS', 'América-RN', 'Amazonas', 'Águia de Marabá', 'Sousa-PB']

serie_a = []
serie_b = []

for i in range(len(times)):
    sorteio = choice(times)

    if i%2 == 0:
        serie_a.append(sorteio)
    else:
        serie_b.append(sorteio)
    times.remove(sorteio)

print(f"\nSerie A:")
for time in serie_a:
    print("-", time)
print(f"\nSerie B:")
for time in serie_b:
    print("-", time)
```
