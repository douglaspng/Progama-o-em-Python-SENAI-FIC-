# Laço de repetição WHILE


```python
# EXEMPLO 01

i = 1

while i <= 10:
    print(i, end=' ')
    i += 1
```

```python
# DESAFIO 01

i = 10

while i >= 1:
    print(i, end=' ')
    i += -1
```

```python
# EXEMPLO 03


i=0
while i < 10:
    i+=1
    if i == 3:
        print("Pulando a iteração 3, voltando para o começo do laço!")
        continue
    print(i)
else:
    print("Fim do loop.")
```

```python
# EXEMPLO 04


while True:
    r = input("Digite (S) para sair ou (C) para continuar\n")
    if r.upper() == 'S':
        print("Saindo do loop...")
        break
    print("\nContinuando...\n")
```

```python
# EXEMPLO 05


while True:
    r = input("\nDigite (S) para sair ou (C) para continuar\n").upper()
    if r != 'S' and r != 'C':
        print("\nDeixa de besteira, só vale S e C! 👶")
        continue
    if r == 'S':
        print("Saindo do loop...")
        break
    print("\nContinuando...\n")
```

```python
# DESAFIO 01
"""
Faça um programa que leia o sexo de uma pessoa, mas só aceite os
valores ‘M’ ou ‘F’. Caso esteja errado, peça a digitação novamente até
ter um valor correto.
"""
from google.colab import output
import time


while True:
    time.sleep(3)
    output.clear()
    r = input("Digite (M) para sair ou (F) para continuar\n").lower()
    if r != 'm' and r != 'f':
        print("\nDeixa de besteira, só vale M e F! ")
        continue
    if r == 'm' or r == 'f':
        output.clear()
        print("Sexo cadastrado! ")
        break
    print("\nSexo Errado ")
```

```python
# DESAFIO 02
"""
Faça um programa onde o computador vai “pensar” em um número entre 1 a 10.
O jogador vai tentar adivinhar até acertar, mostrando no final quantos
palpites foram necessários para vencer.
"""
from google.colab import output
from random import randint

chutes = 1

while True:
    i = float(input("Digite um valor entre 1 e 10 para testar sua sorte "))

    if i == randint(1, 10):
        output.clear()
        print(f"Parabéns, você acertou o número premiado!")
        break
    else:
        output.clear()
        print("\nErrouuu, tente novamente \n")
        chutes += 1
        continue
print(f"\nForam necessárias {chutes} tentativas")
```

```python
# DESAFIO 02 v2


from google.colab import output
import random
pc = random.randint(1, 10)
palpites = 1
errou = False
msg = "\033[31mErrou! 😳\033[0m\n\n"


while True:
    output.clear()
    print(msg if errou else '', end='')
    print(f"Estamos na {palpites}ª rodada, acerte o nº do computador! 💻\n")
    pl = int(input("Digite um nº entre 1 e 10: "))
    if pc == pl:
        output.clear()
        if palpites == 1:
            print(f"\033[32m 👏🏻👏🏻👏🏻 Que sorte! Você venceu em uma tentativa!\033[0m 😁🎆")
        else:
            print(f"\033[32m 👏🏻👏🏻👏🏻 Você venceu em {palpites} tentativas!\033[0m 😁🎆")
        break
    errou = True
    palpites += 1
```

```python
# DESAFIO 03
"""
Crie um programa que leia vários números inteiros pelo teclado. O
programa só vai parar quando o usuário digitar o valor 999, que é a
condição de parada. No final, mostre quantos números foram
digitados e qual foi a soma entre eles (desconsiderando o 999).
"""

chutes = 1
soma_tentativas = 0

while True:
    i = float(input("Digite um valor entre 1 e 1000 para testar sua sorte \n"))
    output.clear()

    if i == 999:
        print(f"Parabéns, você acertou o número premiado!")
        break
    else:
        output.clear()
        print("\nErrouuu, tente novamente \n")
        chutes += 1
        soma_tentativas += i
        continue

print(f"\nForam necessárias {chutes} tentativas e a soma total dos números digitados foram {soma_tentativas}")
```
