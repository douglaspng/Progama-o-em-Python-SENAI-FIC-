```python
# DESAFIO 07
"""
Escreva um código que calcule a hipotenusa de um triângulo retângulo,
cujos catetos são:
a = 4
b = 3
"""
a = 4
b = 3

hipotenusa = a**2 + b**2

valorhipotenusa = hipotenusa** 0.5

print(f"\n Se o valor do primeiro cateto é 4 e o do segundo cateto é 3, a hipotenusa será {valorhipotenusa:.0f}")
```

```python
# DESAFIO 08
"""
Solicite ao usuário o valor do salário atual, em seguida solicite o
percentual de aumento e imprima o valor do salário atualizado.
"""

salario_atual = float(input("\nQual o seu sálario anual\n"))

aumento_porcento = float(input("\nEm um cenário fictício, qual seria o aumento ideal para sua realidade em porcentagem?\n"))

valor_atualizado = salario_atual + salario_atual * (aumento_porcento / 100)

print(f"\nEm um cenário inexistente, seu salário seria {valor_atualizado:,.2f}")
```

# MANIPULAÇÃO DE STRINGS


```python
# ANÁLISE DE STRINGS (str)

n = 'Aula'

print('Numérico:', n.isnumeric())            # Numérico
print('Alfabético:', n.isalpha())            # Alfabético
print('Alfanumérico:', n.isalnum())          # Alfanumérico
print('Maiúscula:', n.isupper())             # Maiúscula
print('Minúscula:', n.islower())             # Minúscula
print('Primeira maiúscula:', n.istitle())    # Primeira maiúscula
print('Numérico:', n.isdecimal())            # Numérico
print('Espaço:', n.isspace())                # Espaço
```

```python
# FATIAMENTO DE STRINGS


texto = "Curso de Python"


print(texto[0])          # 7º caracterer
print(texto[9:15])       # 10º ao 15º caracterer
print(texto[9:15:2])     # 10º ao 15º caracterer de 2 em 2
print(texto[:5])         # Até o 5º caracterer
print(texto[9:])         # 10º caracterer em diante
print(texto[::-1])       # Inverter o texto
print(texto[-1])         # Último caracter
```

```python
# ANÁLISE DE STRINGS (Case sensitive)


frase = "SENAI Ary Torres"


# tamanho da string (quantidade de caracteres)
print(len(frase))


# contar caracteres (ou sequência de caracteres)
print(frase.count('r'))


# encontrar índice (retorna -1 se não encontrar)
print(frase.find(''))


# encontrar índice (a partir da direita)
print(frase.rfind('s'))


# encontrar índice (gera erro se não encontrar)
print(frase.index('e'))


# verificar pertinência
print('ENA' in frase)
```

```python
# TRANSFORMAÇÃO DE STRINGS

# maiúscula
print(frase.upper())

# minúscula
print(frase.lower())

# primeira da frase em maiúscula
print(frase.capitalize())

# primeira letra de cada palavra em maiúscula
print(frase.title())
```

```python
# DESAFIO 05
"""
Refazer o DESAFIO 04 escolhendo a letra para contar e desconsiderar espaços.
"""

palavra_teste = input("Digite qualquer palavra ou frase que você querer: ").lower().replace(' ', '')

letra_escolhida = input("Escolha a letra que você vai querer contar: ").lower()

print(f"A sua letra escolhida, aparece {palavra_teste.count(letra_escolhida)} vezes")
print(f"A primeira posição que a sua letra escolhida aparece é a {palavra_teste.find(letra_escolhida)+1}")
```
