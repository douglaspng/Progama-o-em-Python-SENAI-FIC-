# Tratamento de Erros


```python
# EXEMPLO 01


try:
    numerador   = int(input("Digite o numerador: "))
    denominador = int(input("Digite o denominador: "))
    razao = numerador/denominador
except Exception as erro:
    print(f"Erro encontrado: {erro}")
    print(f"Erro encontrado: {erro.__class__}")
else:
    print(razao)
```

```python
# EXEMPLO ALUNO

try:
    num = int(input("Digite um número: "))
except:
    print("Isso não é um número")
else:
    print("Você digitou corretamente o ", num)
```

```python
# EXEMPLO 02


try:
    # Código que pode gerar erro
    numero = int(input("Digite um número: "))
    resultado = 10 / numero
except(ValueError, TypeError):
    # Trata erros mencionados acima
    print("Digite um número válido")
except(ZeroDivisionError):
    # Trata erro mencionado acima
    print("Não é possível dividir por zero")
else:
    # Executa se não houver erro
    print(f"\nResultado: {resultado}")
finally:
    # Sempre executa
    print("\nEncerrando o programa...")
```

```python
try:
    nota1 = float(input("Digite a 1ª nota: "))
    nota2 = float(input("Digite a 2ª nota: "))
    media = (nota1+nota2)/2
except:
    print("Deu ruim")
else:
    if media < 5:
        print(f"\nMédia: {media:,.1f}")
        print("Resultado: REPROVADO")
    elif media < 7:
        print(f"\nMédia: {media:,.1f}")
        print("Resultado: RECUPERAÇÃO")
    else:
        print(f"\nMédia: {media:,.1f}")
        print("Resultado: APROVADO")
```

```python
# EXEMPLO 04


import time
from google.colab import output


while True:
    try:
        output.clear()
        nota1 = float(input("Digite a 1ª nota: "))
        nota2 = float(input("Digite a 2ª nota: "))
        media = (nota1+nota2)/2
    except:
        print("Deu ruim")
        time.sleep(3)
        continue
    else:
        if media < 5:
            print(f"\nMédia: {media:,.1f}")
            print("Resultado: REPROVADO")
        elif media < 7:
            print(f"\nMédia: {media:,.1f}")
            print("Resultado: RECUPERAÇÃO")
        else:
            print(f"\nMédia: {media:,.1f}")
            print("Resultado: APROVADO")
        break
```

```python
# DESAFIO 01


aluno = {}

try:
    aluno['nome'] = input("Digite o nome do aluno: ")
    aluno['media'] = float(input("Digite a média do aluno: "))
except Exception as erro:
    print(f"Ocorreu um erro ({erro.__class__}), preencha novamente... ")
    time.sleep(2)
else:
    if aluno['media'] < 5:
        aluno['situacao'] = "Reprovado"
    elif aluno['media'] < 7:
        aluno['situacao'] = "Exame"
    else:
        aluno['situacao'] = "Aprovado"

    print(f"\nNome: {aluno['nome']}")
    print(f"Média: {aluno['media']}")
    print(f"Situação: {aluno['situacao']}")

```
