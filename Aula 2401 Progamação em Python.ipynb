ALGORITIMO LANCHE
-
Passo 1: Localize os ingredientes, Utencílios e Eletrodomésticos necessários:

Ingredientes: Baguete de 3 queijos, 3 folhas alface, 1 tomate, 150g frango (pré cozido), 150g bacon (selado), cream chease, manteiga, molho barbecue;

Utencílios: Faca, colher, papel manteiga, prato, bandeija de metal e frigideira;

Eletrodomésticos: Fogão com Forno.

-
OBS: O Frango e bacon deve estarem pré prontos antes da receita.

-
Passo 2: Pegue a baguete de 3 queijos, abra com a faca e passe a manteiga e sele na frigideira;

Passo 3: Corte o tomate em fatias, pegue as 3 fatias de alface e junto com o tomate lave-os;

Passo 4: Retire o baguete de 3 queijos selado com manteiga da frigideira após 5m;

Passo 5: coloque a baguete de 3 queijos selado em cima da bandeija de metal, dentro da baguete espalhe o frango pré-cozido e o bacon pré-frito;

Passo 6: Coloque a bandeija com abaguete de 3 queijos selada com frango e bacon no forno pré aquecido a 180°, e deixe por aproximadamente 10min no forno.

Passo 7: Retire a bandeija com a baguete 3 queijos selada com frango e bacon do forno;

Passo 8: Em cima de uma mesa coloque o papel manteiga com 30cm de altura e 30cm de comprimento;

Passo 9: Coloque a baguete com os outros ingredientes em cima do papel manteiga;

Passo 10: Adicione as 3 folhas de alface lavada e o tomate cortado dentro da baguete, adicione o molho barbecue e cream chease a gosto.

Passo 11: Aproveite seu lanche!.

```python
# Operadores aritméticos


a = 10
b = 3


print(a +  b)     # 13      (adição)
print(a -  b)     # 7       (subtração)
print(a *  b)     # 30      (multiplicação)
print(a /  b)     # 3.3333  (divisão)
print(a // b)     # 3       (divisão inteira)
print(a %  b)     # 1       (módulo/resto da divisão)
print(a ** b)     # 1000    (exponenciação)
print(a ** (1/b)) # 2.14    (radiciação) (a ** 0.5)
```

```python
# Operadores relacionais ou de comparação (if | elif | while)


a = 10
b = 3


print(a == b)  # False  Igual
print(a != b)  # True   Diferente
print(a >  b)  # True   Maior
print(a <  b)  # False  Menor
print(a >= b)  # True   Maior ou igual
print(a <= b)  # False  Menor ou igual
```

```python
# Operadores lógicos (e conectores lógicos and/or)


a = True
b = False


print(a and b)  # False
print(a or b)   # True
print(not a)    # False
```

```python
# Operadores de atribuição


a = 10
a += 2  # a = a + 2
print(a) # 12


b = 10
b -= 2  # b = b - 2
print(b) # 8


c = 10
c *= c  # c = c * c
print(c) # 100


d = 10
d /= 2 # d = d / 2
print(d) # 5.0


d = 10
d **= 2 # d = d ** 2
print(d) # 100


d = 10
d //= 2
print(d)


d = 10
d %= 2
print(d) # 0
```
