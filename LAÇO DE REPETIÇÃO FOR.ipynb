# LAÇO DE REPETIÇÃO FOR

```python
# EXEMPLO 01 (i de iteração)

for i in range(1, 11):
    print(i, end=' ')
```

```python
# EXEMPLO 02 (Contagem Regressiva)

for i in range(10, 0, -1):
    print(i, end=' ')
```

```python
# EXEMPLO 02 v2


inicio, fim, salto = 5, 0, -1


for i in range(inicio, fim, salto):
    print(i, end=' ')
```

```python
# EXEMPLO 03
# range(início, fim, salto)


for i in range(1, 10, 2):
    print(i, end=' ')
```

```python
# EXEMPLO 04


for i in range(3):
    print(f"{i+5} | SENAI")
```

```python
# EXEMPLO 05


texto = 'Python'


for letra in texto:
    # pass
    print('\033[31m', letra.upper(), end='   ')
```

```python
# EXEMPLO 06

frase = ['PYTHON', 'É', 'FANTÁSTICO']

for palavra in frase:
    print(palavra, end='\t')
```

```python
# EXEMPLO 06 (DESAFIO)

frase = ['PYTHON', 'É', 'FANTÁSTICO']

for palavra in frase:
    if palavra != 'PYTHON':
        print(palavra)
```

```python
# EXEMPLO 07
""" continue ignora o resto da iteração """


lista = [1, 10, 20, 30, 40, 50]


for numero in lista:
    if numero == 30:
        continue
    print(numero, end=' ')
```

```python
# EXEMPLO 08
""" break finaliza o laço """


for i in range(50):
    if i >= 10:
        break
    print(i+1, end=' ')
```

```python
# EXEMPLO 09

frase = ['PYTHON', 'É', 'FANTÁSTICO!']


for i in range(len(frase)):
    print(frase[i])
```

```python
# EXEMPLO 10

frase = ['PYTHON', 'É', 'FANTÁSTICO!']


for i, palavra in enumerate(frase):
    print(i, palavra)
```

```python
# EXEMPLO PRINT SEP


lista = [10, 20, 30 , 40]


for i, numero in enumerate(lista):
    print(i, numero, sep='/')
```

```python
# EXEMPLO SEP PRINT

dd = '02'
mm = '12'
aa = '25'

print(dd, mm, aa, sep='/')
```

```python
# EXEMPLO 11

for i in range(1, 11):
    for j in range(1, 11):
        print(f"{i*j}".rjust(3), end=' ')
    print()
```

```python
# EXEMPLO 12

for i in range(0, 11):
    for j in range(0, 11):
        print(f"{i+j}".rjust(3), end=' ')
    print()
```
